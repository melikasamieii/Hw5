```python
import pandas as pd

df=pd.read_csv("iris_dataset.csv")

df.head()
```

```python
x= df[["sepal length (cm)", "sepal width (cm)" ]].values
#label
y = df["target"].values
```

```python
ptest = x[149]
ytest = y[149]

ptrain = x[:149]
ytrain = y[:149]

ytest , ptest, ptrain
```

```python
import numpy as np

def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))

distances = []

for i in range(len(ptrain)):
    d = euclidean_distance(ptest, ptrain[i])
    distances.append((d, ytrain[i]))
distances    
```

```python
def cldistance(item):
    return item[0]
distances = sorted(distances , key= cldistance)

distances

# distances = sorted(distances , key= lambda x:x[0])
```

```python
k= 1
nghbrk1 = distances[:k]
predk1 = nghbrk1[0][1]
#dist[0] + class[1]
k = 3
nghbrk3 = distances[:k]
classes = [n[1] for n in nghbrk3]
#n[1] = classes in tuple

predk3 = max(set(classes), key= classes.count)    


k = 100
nghbrk100 = distances[:k]
classes = [n[1] for n in nghbrk3]
#n[1] = classes in tuple

predk100 = max(set(classes), key= classes.count) 

ytest, predk1, predk3 ,predk100
```
