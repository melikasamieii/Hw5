#1

نفرین بعدیت یکی از اساسی‌ترین چالش‌های یادگیری ماشین و به‌ویژه در مسائل یادگیری با نظارت است که با افزایش تعداد ویژگی‌ها (dimensions) در یک مجموعه‌داده، حجم فضای ویژگی به‌صورت نمایی افزایش می‌یابد، در حالی که تعداد نمونه‌های موجود ثابت است. نتیجه این امر «رقیق شدن داده‌ها» (data sparsity) در فضای ویژگی است.

۲. اثر افزایش و کاهش بُعد بر فاصلهٔ اقلیدسی

وقتی تعداد ویژگی‌ها (n) زیاد می‌شود:

الف) فاصله‌ها به‌طور یکنواخت بزرگ می‌شوند
در نتیجه:

نقطه‌ها از هم «دورتر» ظاهر می‌شوند

مدل‌هایی مثل KNN سخت‌تر می‌توانند نزدیک‌ترین همسایه را شناسایی کنند

نسبت فاصله نزدیک‌ترین و دورترین نقطه به یکدیگر نزدیک می‌شود

ب) کاهش تمایز بین نقاط

در ابعاد بالا، تفاوت‌ها از بین می‌روند.
برای مثال:

در بُعد ۲، فاصله‌ها تنوع زیادی دارند

در بُعد ۱۰۰، تقریباً تمام فاصله‌ها مشابه‌اند
این باعث می‌شود مدل در تشخیص مرزهای واقعی ناتوان شود.

۳. اثر کاهش بُعد

وقتی ویژگی‌های اضافی حذف شوند:

فاصله‌ها دوباره معنا‌دار می‌شوند

مدل بهتر نقطه‌های مشابه را تشخیص می‌دهد

احتمال Overfitting کم می‌شود

سرعت محاسبات به‌شدت بیشتر می‌شود

۴. ابزارهای کاهش اثر COD

الف) کاهش بُعد (Dimensionality Reduction)

PCA → تبدیل ویژگی‌ها به مؤلفه‌های اصلی

t-SNE / UMAP → نگاشت داده‌ها به ابعاد کم‌تر با حفظ ساختار

LDA → کاهش بُعد نظارت‌شده

ب) انتخاب ویژگی (Feature Selection)

انتخاب مهم‌ترین فیچرها

حذف نویز

روش‌های آماری یا یادگیری

ج) نرمال‌سازی و اسکیلینگ

کمک می‌کند اثر ابعاد ناهماهنگ کم شود.

نشستن

ایستادن

غذا خوردن

دویدن

هر سنسور یک ویژگی است. افزایش تعداد سنسورها دقت را بالا می‌برد اما هزینه را افزایش می‌دهد.

 مهم‌ترین بخش برای تشخیص رفتار
حرکت پا مشخص‌ترین شاخص فعالیت است.
برای مثال:

ایستاده فشار روی ۴ پا

راه رفتن  جابه‌جایی متناوب پاها

خوابیدن  عدم فعالیت پاها و تغییر زاویه

حداقل 2 یا 4 سنسور در پاها ضروری هستند.

ب) سر و گردن

برای رفتارهای خاص لازم است:

غذا خوردن (حرکت زیاد سر)

نوشیدن

نگاه کردن به اطراف

ج) بخش شكم / پهلو

برای تشخیص:

خوابیدن

نشستن

حداقل تعداد سنسور: ۳ تا ۵ عدد
مکان پیشنهادی:

یک سنسور روی گردن – تشخیص غذا خوردن / نوشیدن / حرکت بدن

دو سنسور روی دو پای جلو – کافی برای تحلیل راه رفتن / ایستادن / نشستن

یک سنسور روی پهلو – تشخیص خوابیدن / استراحت

#3

تیم امنیت: جلوگیری از تقلب با استفاده از میانگین موجودی حساب
می‌توان از الگوهای موجودی حساب  استفاده کرد:

مشتریان واقعی نوسان طبیعی دارند

کلاهبرداران معمولاً الگوهای ناگهانی دارند

افزایش/کاهش شدید موجودی ممکن است نشانه باشد

مقایسه رفتار فرد با خود فرد در گذشته (Behavioral Modeling)

روش‌ها:

استفاده از Z-Score برای تشخیص رفتار غیرعادی

شناسایی تراکنش‌هایی که انحراف زیادی از میانگین دارند

بررسی تراکنش‌های پی‌درپی غیرعادی

گام ۲ – تیم پروفشنال: دور ماندن از سیستم امنیتی

کسی که قصد پولشویی دارد:

الگوهای طبیعی کاربر را تقلید می‌کند

مبلغ تراکنش را کمی بالاتر یا کمتر از حد حساسیت انتخاب می‌کند

تراکنش‌ها را در بازه‌های زمانی مشابه گذشته انجام می‌دهد

به‌جای یک تراکنش بزرگ، چند تراکنش کوچک انجام می‌دهد
(Structuring / Smurfing)

گام ۳ – تیم امنیت: شناسایی افراد مخفی

تیم امنیت برای مقابله:

شناسایی رفتارهای خردشده (Small Frequent Transactions)

بررسی مجموع تراکنش‌ها در بازهٔ ثابت

بررسی ارتباط حساب‌های متعدد

استفاده از مدل‌های sequence-based مانند LSTM برای الگوگیری

شناسایی دستگاه / مکان / IP مشابه

گام ۴ – تیم پروفشنال: چطور گمنام بمانیم؟

یک مهاجم:

از چند حساب مختلف با هویت‌های متفاوت استفاده می‌کند

تراکنش‌ها را از مکان‌های مختلف می‌فرستد (VPN / پراکندگی مکانی)

رفتار گذشتهٔ کاربر واقعی را شبیه‌سازی می‌کند

تراکنش‌ها را در ساعاتی مشابه رفتار عادی انجام می‌دهد

گام ۵ – اگر امنیت فقط از Range داده‌ها استفاده کند؟

مشکل بسیار جدی ایجاد می‌شود:

دامنه (min–max) فقط محدوده را نشان می‌دهد، نه الگوها

رفتارهای جزئی غیرعادی شناسایی نمی‌شوند

مهاجم می‌تواند داخل Range بماند و دیده نشود

نوعی Blind Spot ایجاد می‌شود

بنابراین استفادهٔ تنها از Range، مدل را بسیار ساده و آسیب‌پذیر می‌کند.
