```python
import pandas as pd

df=pd.read_csv("iris_dataset.csv")

df.head()
```

```python
x= df.iloc[:, :4].values
#label
y = df["target"].values
```

```python
ptest = x[149]
ytest = y[149]

ptrain = x[:149]
ytrain = y[:149]

ytest , ptest, ptrain
```

```python
import numpy as np

def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))

distances = []

for i in range(len(ptrain)):
    d = euclidean_distance(ptest, ptrain[i])
    distances.append((float(d), int(ytrain[i])))

def cldistance(item):
    return item[0]
distances = sorted(distances , key= cldistance)    
```

```python
k = 3
nghbrk3 = distances[:k]
classes = [n[1] for n in nghbrk3]
#n[1] = classes in tuple

predk3 = max(set(classes), key= classes.count)  

ytest, distances[:3] , predk3
```

```python
mean = ptrain.mean(axis=0)
std = ptrain.std(axis=0)
ptrain_norm = (ptrain- mean)/std
ptest_norm = (ptest- mean)/ std
ptest_norm.round(4)
```

```python
distances_norm = []

for i in range(len(ptrain_norm)):
    d = euclidean_distance(ptest_norm, ptrain_norm[i])
    distances_norm.append((d, ytrain[i]))

  
def cldistance_norm(item):
    return item[0]
distances_norm = sorted(distances_norm , key= cldistance_norm)  
```

```python
k = 3
nghbrk3 = distances_norm[:k]
classes = [n[1] for n in nghbrk3]
#n[1] = classes in tuple

predk3_norm = max(set(classes), key= classes.count)  

distances_norm[:3], predk3_norm, predk3, ytest
```
