```python
import pandas as pd

df=pd.read_csv("iris_dataset.csv")

df.head()





```

```python
mean=df.groupby("target").mean()
mean
```

```python
std=df.groupby("target").std()
std
```

```python
maxi=df.groupby("target").max()
mini=df.groupby("target").min()
anomaly = maxi - mini
anomaly
```

```python
columns = ["sepal length (cm)",	"sepal width (cm)",	"petal length (cm)","petal width (cm)"]
corrmatrix = df[columns].corr()
corrmatrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9,8))
sns.heatmap(corrmatrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
corrdrop = corrmatrix.unstack()
corrdrop = corrdrop[corrdrop != 1]
maxcorr = corrdrop.idxmax()
value = corrdrop.max()

maxcorr[0], maxcorr[1] , value

```
