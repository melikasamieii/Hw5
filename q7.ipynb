```python
import pandas as pd

df=pd.read_csv("iris_dataset.csv")

df.head()
```

```python
x = df.iloc[:, :4].values
y= df.iloc[:, 4].values
```

```python
class_df = df.groupby("target")

mean = class_df.mean()
mini = class_df.min()
maxi = class_df.max()
vari = class_df.var()
rangedf = maxi - mini
mean
```

```python
import numpy as np
def gauss(x , mu,var):
    sig = np.sqrt(var)
    return (1/(np.sqrt(2*np.pi)*sig))*np.exp(-0.5*((x-mu)/sig)**2)

```

```python
petal = df["petal length (cm)"].values
y_t = df["target"].values
class_ids = mean.index
class_ids
```

```python

mu = mean["petal length (cm)"]
var = vari["petal length (cm)"]
print(mu , var)
```

```python
def calculateLabel(petal):
    y_pred = []
    for x in petal:
        probs = {}

        for c in class_ids:
            probs[c] = gauss(x, mu[c] , var[c])

        bestcls = max(probs , key= probs.get)
        y_pred.append(bestcls)
    return y_pred
```

```python
correct = 0
y_pred = calculateLabel(petal)
for i in range (len(y_t)):
    correct+= y_pred[i] == y_t[i]
print(correct)  
df.info()  
```

```python

def Accuracy(correct_answer, total):
    precision = (correct_answer / total)*100
    return precision
```

```python
accuracy = Accuracy(correct, len(y_t))
accuracy
```

```python
mu_sep_len  = mean["sepal length (cm)"]
var_sep_len = vari["sepal length (cm)"]

mu_sep_wid  = mean["sepal width (cm)"]
var_sep_wid = vari["sepal width (cm)"]

mu_pet_len  = mean["petal length (cm)"]
var_pet_len = vari["petal length (cm)"]

mu_pet_wid  = mean["petal width (cm)"]
var_pet_wid = vari["petal width (cm)"]
```

```python
sep_len   = df["sepal length (cm)"].values
sep_wid   = df["sepal width (cm)"].values
pet_len   = df["petal length (cm)"].values
pet_wid   = df["petal width (cm)"].values
y_t       = df["target"].values
```

```python


sep_len_test = test["sepal_length"].values
sep_wid_test = test["sepal_width"].values
pet_len_test = test["petal_length"].values
pet_wid_test = test["petal_width"].values
y_testt      = test["label"].values
```

```python
def calculateLabel_4feat(sep_len, sep_wid, pet_len, pet_wid):
    y_pred = []
    
    for i in range(len(sep_len)):
        x1 = sep_len[i]   
        x2 = sep_wid[i]  
        x3 = pet_len[i]   
        x4 = pet_wid[i]  

        probs_classes = {} 

        for c in class_ids:  
            P1 = gauss(x1, mu_sep_len[c],  var_sep_len[c])
            P2 = gauss(x2, mu_sep_wid[c],  var_sep_wid[c])
            P3 = gauss(x3, mu_pet_len[c],  var_pet_len[c])
            P4 = gauss(x4, mu_pet_wid[c],  var_pet_wid[c])

            probs_feat = [P1, P2, P3, P4]
            top2 = sorted(probs_feat, reverse=True)[:2]   
            P_final = sum(top2) / 2                       
            probs_classes[c] = P_final

        bestcls = max(probs_classes, key=probs_classes.get)
        y_pred.append(bestcls)

    return y_pred
```

```python
y_pred_train_4 = calculateLabel_4feat(sep_len, sep_wid, pet_len, pet_wid)

correct_train_4 = 0
for i in range(len(y_t)):
    correct_train_4 += (y_pred_train_4[i] == y_t[i])

train_acc_4 = Accuracy(correct_train_4, len(y_t))
print("Train accuracy (4 features, top2) =", train_acc_4)
```

```python
y_pred_test_4 = calculateLabel_4feat(sep_len_test, sep_wid_test, pet_len_test, pet_wid_test)

correct_test_4 = 0
for i in range(len(y_testt)):
    correct_test_4 += (y_pred_test_4[i] == y_testt[i])

test_acc_4 = Accuracy(correct_test_4, len(y_testt))
print("Test accuracy (4 features, top2) =", test_acc_4)
```
